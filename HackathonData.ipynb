#### Code to match user to food

```python
from enum import Enum, auto
from typing import List, Tuple

class ChronicDisease(Enum):
    DIABETES = auto()
    HYPERTENSION = auto()
    HEART_DISEASE = auto()
    NONE = auto()

class MedicalCondition(Enum):
    CANCER = auto()
    KIDNEY_DISEASE = auto()
    RESPIRATORY_ILLNESS = auto()
    NONE = auto()

class Allergy(Enum):
    NUTS = auto()
    DAIRY = auto()
    GLUTEN = auto()
    SEAFOOD = auto()
    NONE = auto()

class DietaryRestriction(Enum):
    VEGETARIAN = auto()
    VEGAN = auto()
    HALAL = auto()
    NO_BEEF = auto()
    NO_PORK = auto()
    NONE = auto()

class FoodType(Enum):
    NON_PERISHABLE = auto()
    FRESH_PRODUCE = auto()
    DAIRY = auto()
    MEAT = auto()
    SEAFOOD = auto()
    HALAL = auto()
    VEGETARIAN = auto()
    VEGAN = auto()
    GLUTEN_FREE = auto()

class Recipient:
    def init(self, name, age, chronic_diseases, medical_conditions, allergies, family_size, income_level, employment_status, dietary_restrictions, location):
        self.name = name
        self.age = age
        self.chronic_diseases = chronic_diseases
        self.medical_conditions = medical_conditions
        self.allergies = allergies
        self.family_size = family_size
        self.income_level = income_level
        self.employment_status = employment_status
        self.dietary_restrictions = dietary_restrictions
        self.location = location

class Donor:
    def init(self, name, food_types, quantity):
        self.name = name
        self.food_types = food_types
        self.quantity = quantity

def classify_recipient(recipient):
    points = 0
    
    # Chronic diseases (0-30 points)
    if ChronicDisease.NONE not in recipient.chronic_diseases:
        points += 30 * len(recipient.chronic_diseases)
    
    # Medical conditions (0-25 points)
    if MedicalCondition.NONE not in recipient.medical_conditions:
        points += 25 * len(recipient.medical_conditions)
    
    # Allergies (0-20 points)
    if Allergy.NONE not in recipient.allergies:
        points += 20 * len(recipient.allergies)
    
    # Age
    if recipient.age >= 65 or recipient.age <= 12:
        points += 15
    elif 13 <= recipient.age <= 18:
        points += 10
    
    # Income level
    if recipient.income_level == 'low':
        points += 10
    elif recipient.income_level == 'medium':
        points += 5
    
    # Employment status
    if recipient.employment_status == 'unemployed':
        points += 8
    elif recipient.employment_status == 'part-time':
        points += 5
    
    # Dietary restrictions
    if DietaryRestriction.NONE not in recipient.dietary_restrictions:
        points += 5 * len(recipient.dietary_restrictions)
    
    # Location
    if recipient.location == 'far':
        points += 5
    elif recipient.location == 'medium':
        points += 3
    
    vulnerability_level = 'High Vulnerability' if points >= 90 else 'Medium Vulnerability' if points >= 60 else 'Low Vulnerability'
    return vulnerability_level, points

def is_compatible_food(donor_food_types, recipient):
    for food_type in donor_food_types:
        if food_type == FoodType.SEAFOOD and Allergy.SEAFOOD in recipient.allergies:
            return False
        if food_type == FoodType.DAIRY and Allergy.DAIRY in recipient.allergies:
            return False
        if food_type == FoodType.MEAT and DietaryRestriction.VEGETARIAN in recipient.dietary_restrictions:
            return False
        if food_type not in [FoodType.VEGETARIAN, FoodType.VEGAN, FoodType.HALAL] and DietaryRestriction.HALAL in recipient.dietary_restrictions:
            return False
    return True

def match_donations_with_priority(donors: List[Donor], recipients: List[Recipient]) -> Tuple[List[Tuple[Donor, Recipient, str]], List[Recipient]]:
    matches = []
    recipients_with_points = [(r, *classify_recipient(r)) for r in recipients]
    recipients_with_points.sort(key=lambda x: x[2], reverse=True)
    
    while recipients_with_points:
        recipient, vulnerability_level, points = recipients_with_points.pop(0)
        compatible_donors = [d for d in donors if is_compatible_food(d.food_types, recipient)]
        
        if compatible_donors:
            # Choose the donor with the highest quantity
            best_donor = max(compatible_donors, key=lambda d: d.quantity)
            
            priority = "High Priority" if points >= 90 else "Medium Priority" if points >= 60 else "Low Priority"
            matches.append((best_donor, recipient, priority))
            
            # Remove the donor after matching
            donors.remove(best_donor)
    
    # The remaining recipients are unmatched
    unmatched_recipients = [r for r, _, _ in recipients_with_points]
    
    return matches, unmatched_recipients


# Example usage
recipients = [
    Recipient("Alice", 70, [ChronicDisease.DIABETES], [MedicalCondition.NONE], [Allergy.NONE], 1, "low", "unemployed", [DietaryRestriction.HALAL], "far"),
    Recipient("Bob", 45, [ChronicDisease.NONE], [MedicalCondition.CANCER], [Allergy.SEAFOOD], 4, "medium", "part-time", [DietaryRestriction.NO_BEEF], "close"),
    Recipient("Charlie", 10, [ChronicDisease.NONE], [MedicalCondition.NONE], [Allergy.NUTS, Allergy.GLUTEN], 5, "low", "unemployed", [DietaryRestriction.VEGETARIAN], "medium"),
    Recipient("David", 30, [ChronicDisease.NONE], [MedicalCondition.NONE], [Allergy.NONE], 2, "medium", "full-time", [DietaryRestriction.NONE], "close"),
    Recipient("Eve", 50, [ChronicDisease.HYPERTENSION], [MedicalCondition.NONE], [Allergy.DAIRY], 3, "medium", "part-time", [DietaryRestriction.NONE], "medium"),
    Recipient("Frank", 68, [ChronicDisease.HEART_DISEASE, ChronicDisease.DIABETES], [MedicalCondition.NONE], [Allergy.NONE], 1, "low", "unemployed", [DietaryRestriction.HALAL], "far"),
    Recipient("Grace", 12, [ChronicDisease.NONE], [MedicalCondition.KIDNEY_DISEASE], [Allergy.NUTS], 5, "low", "unemployed", [DietaryRestriction.VEGETARIAN], "medium"),
    Recipient("Henry", 72, [ChronicDisease.DIABETES, ChronicDisease.HEART_DISEASE], [MedicalCondition.KIDNEY_DISEASE], [Allergy.DAIRY], 1, "low", "unemployed", [DietaryRestriction.HALAL], "far"),
    Recipient("Isabel", 68, [ChronicDisease.HYPERTENSION, ChronicDisease.DIABETES], [MedicalCondition.CANCER, MedicalCondition.RESPIRATORY_ILLNESS], [Allergy.GLUTEN, Allergy.NUTS], 2, "low", "unemployed", [DietaryRestriction.VEGETARIAN, DietaryRestriction.NO_PORK], "medium"),
    Recipient("Jack", 15, [ChronicDisease.NONE], [MedicalCondition.CANCER], [Allergy.SEAFOOD, Allergy.NUTS], 5, "low", "unemployed", [DietaryRestriction.HALAL, DietaryRestriction.NO_PORK], "far"),
    Recipient("Karen", 10, [ChronicDisease.NONE], [MedicalCondition.RESPIRATORY_ILLNESS], [Allergy.DAIRY, Allergy.GLUTEN], 4, "low", "unemployed", [DietaryRestriction.VEGETARIAN], "close"),
    Recipient("Laura", 75, [ChronicDisease.HEART_DISEASE, ChronicDisease.HYPERTENSION], [MedicalCondition.KIDNEY_DISEASE], [Allergy.NONE], 2, "low", "unemployed", [DietaryRestriction.NONE], "far")
]

donors = [
    Donor("Donor1", [FoodType.NON_PERISHABLE, FoodType.GLUTEN_FREE], 100),
    Donor("Donor2", [FoodType.FRESH_PRODUCE, FoodType.VEGETARIAN], 50),
    Donor("Donor3", [FoodType.HALAL, FoodType.MEAT], 75),
    Donor("Donor4", [FoodType.FRESH_PRODUCE, FoodType.DAIRY], 30),
    Donor("Donor5", [FoodType.NON_PERISHABLE, FoodType.SEAFOOD], 50),
    Donor("Donor6", [FoodType.VEGETARIAN, FoodType.GLUTEN_FREE], 40),
    Donor("Donor7", [FoodType.HALAL, FoodType.MEAT], 50),
    Donor("Donor8", [FoodType.NON_PERISHABLE, FoodType.HALAL, FoodType.GLUTEN_FREE], 120),
    Donor("Donor9", [FoodType.FRESH_PRODUCE, FoodType.VEGETARIAN, FoodType.GLUTEN_FREE], 100),
    Donor("Donor10", [FoodType.HALAL, FoodType.MEAT, FoodType.SEAFOOD], 150),
    Donor("Donor11", [FoodType.VEGAN, FoodType.FRESH_PRODUCE, FoodType.HALAL], 80),
    Donor("Donor12", [FoodType.NON_PERISHABLE, FoodType.DAIRY, FoodType.MEAT], 90)
]

matches, unmatched_recipients = match_donations_with_priority(donors, recipients)

# Printing matches
for donor, recipient, priority in matches:
    print(f"Match: {donor.name} -> {recipient.name} ({priority})")
    print(f"  Donor food types: {[ft.name for ft in donor.food_types]}")
    print(f"  Recipient dietary restrictions: {[dr.name for dr in recipient.dietary_restrictions]}")
    print(f"  Recipient allergies: {[a.name for a in recipient.allergies]}")
    print()

# Print unmatched donors and recipients
unmatched_donors = [d.name for d in donors]

if unmatched_donors:
    print("Unmatched Donors:")
    for donor_name in unmatched_donors:
        print(f"  {donor_name}")

if unmatched_recipients:
    print("Unmatched Recipients:")
    for recipient_name in unmatched_recipients:
        print(f"  {recipient_name}")
```
